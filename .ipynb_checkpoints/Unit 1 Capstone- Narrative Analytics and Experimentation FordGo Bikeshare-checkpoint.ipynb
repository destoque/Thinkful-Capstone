I decided to examine which days of the week were the least popular for riders and why. The days are labeld 0-6 with 0 being Monday and 6 being Sunday. The chart and table below show that Saturdays and Sundays are the least popular days. The most popular days are Monday, Tuesday, and Wednesday, while the least popular days are Saturday and Sunday. 

```python
plt.figure(figsize=(20, 10))


#creating a dataframe that shows the number of bikes rented by day of the week
df_day_data['day'] = pd.DatetimeIndex(df_day_data['start_time']).dayofweek

day_counts= df_day_data['day'].value_counts()

#adding columns to new df
day_df = pd.DataFrame({'day':day_counts.index ,'number':day_counts.values})
day_df

sns.barplot(x='day', y='number', data=day_df)

#demonstrating rider counts by day in a table
df_day_data.day.value_counts()

#sns.barplot(df_day_data.day.value_counts())
```

I then worked to see why that my be by looking at the most popular times of day. The most popular times of day were commuting hours between 6am and 8am and 4pm and 6 pm. This leads me to believe that the reason to Saturday and Sunday are the least popular days is that the bikes are used to commute to work.  

```python
plt.figure(figsize=(20, 10))

df['time_of_day'] = pd.DatetimeIndex(df['start_time']).hour
df['time_of_day']

plt.hist(df['time_of_day'], bins=24, color='red')


plt.ylabel('Number of bike rentals', fontsize=15)
plt.xlabel('Hour of the day', fontsize=15)
plt.title('Most Frequent Time of Day Bikes are Rented', fontsize=20)
plt.xticks(range(0, 24))
plt.show()

df.time_of_day.value_counts()
```

### The Problem
As deciphered above the number of trips is concentrated to subscribers who commute to work on Mondays, Tuesdays, and Wednesdays. Saturdays and Sundays combined only account for 18% of the week's total trips. The goal is to increase ridership on Saturdays and Sundays to turn a higher profit on the weekends.

## Increasing Ridership on the Weekends
Currently, single ride passes are 2.00 for a single use ride, 10 for a day pass, and 15 for a monthly membership. The FordGO bikeshare prices will be reduced on weekends to promote the use of the bikes. 

Prices will be reduced to 1.50 for a single ride pass and a weekend pass will cost users 10 for unlimited rides throughout the weekend. Monthly memberships will remain the same.

The new pricing program wil be tested at all bikeshare stations on weekends for the month of July.

## New Pricing Program Evaluation Metrics

The ridership and profits will be judged against both the 12 month average and the data collected from July 2017 to see if ridership and profits increased. We will use the mean aggregate for the passed 12 months as a benchmark but, will also need to evaluate July because riding habits may be different in the month of July do to an increase in summer activities.

There may be a situation where ridership increases but, profit does not. In that case, we could adjust the prices to return more profit for FordGo Bikeshare
